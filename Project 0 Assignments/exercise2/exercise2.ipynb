```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def fourier_transform(x):
    N = x.size
    n = np.arange(N)
    k = n.reshape((N, 1))
    
    e = np.exp(-2j * np.pi * k * n / N)
    X_k = np.zeros_like(k, dtype=np.complex128)
    
    for row in n:
        for col in n:
            X_k[row] += e[row, col] * x[col]
    return X_k


def func(x):
    return np.exp(-np.pi * x**2)



x = func(np.linspace(-10, 10, 1000))
plt.plot(x)

plt.plot(fourier_transform(x))
```
