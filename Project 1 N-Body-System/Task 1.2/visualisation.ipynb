```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
files = [
    [0, 'output/data/direct_force_0_softening.txt', 'output/graphs/direct_force_0.pdf'],
    [0.001, 'output/data/direct_force_0001_softening.txt', 'output/graphs/direct_force_0001.pdf'],
    [0.01, 'output/data/direct_force_001_softening.txt', 'output/graphs/direct_force_001.pdf'],
    [0.1, 'output/data/direct_force_01_softening.txt', 'output/graphs/direct_force_01.pdf'],
    [1, 'output/data/direct_force_1_softening.txt', 'output/graphs/direct_force_1.pdf'],
    [10, 'output/data/direct_force_100_softening.txt', 'output/graphs/direct_force_10.pdf'],
    [100, 'output/data/direct_force_100_softening.txt', 'output/graphs/direct_force_100.pdf'],
]
```

```python
# generate the overview image for each softening value standalone
if False:
    for softening, filename, export_name in files:
        values = np.genfromtxt(filename, delimiter=',')

        # compute the radius
        radii = [np.linalg.norm(a) for a in values[:, [0,1,2]]]
        abs_forces = [np.linalg.norm(a) for a in values[:, [3,4,5]]]

        plt.scatter(radii, abs_forces, s=5, label=f'softening = {softening}')
        plt.loglog()
        plt.xlabel('Distance to center [a.u.]')
        plt.ylabel('Absolute force value [a.u.]')
        plt.title(f'Direct Force Calculation | softening = {softening}')
        plt.savefig(export_name, dpi=1000)
        plt.close()
```

```python
# generate the overview image of all softening values
if False:
    for softening, filename, export_name in files:
        values = np.genfromtxt(filename, delimiter=',')

        # compute the radius
        radii = [np.linalg.norm(a) for a in values[:, [0,1,2]]]
        abs_forces = [np.linalg.norm(a) for a in values[:, [3,4,5]]]

        plt.scatter(radii, abs_forces, s=2, label=f'softening = {softening}', alpha=0.1)
    plt.loglog()
    plt.xlabel('Distance to center [a.u.]')
    plt.ylabel('Absolute force value [a.u.]')
    plt.title(f'Direct Force Calculation | softening = {softening}')
    plt.legend()
    plt.savefig('output/graphs/direct_force_all.pdf', dpi=1000)
    plt.close()
```
