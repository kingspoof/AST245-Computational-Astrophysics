```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
files = [
    [0, 'output/data/direct_force_0_softening.txt', 'output/graphs/direct_force_0.png'],
    [0.001, 'output/data/direct_force_0001_softening.txt', 'output/graphs/direct_force_0001.png'],
    [0.01, 'output/data/direct_force_001_softening.txt', 'output/graphs/direct_force_001.png'],
    [0.1, 'output/data/direct_force_01_softening.txt', 'output/graphs/direct_force_01.png'],
    [1, 'output/data/direct_force_1_softening.txt', 'output/graphs/direct_force_1.png'],
    [10, 'output/data/direct_force_100_softening.txt', 'output/graphs/direct_force_10.png'],
]
```

```python
# generate the overview image for each softening value standalone
hq_values = np.genfromtxt("output/data/hq_forces.txt", delimiter=',')
hq_radii = [np.linalg.norm(a) for a in hq_values[:, [0,1,2]]]

order = np.argsort(hq_radii)

hq_radii = np.array(hq_radii)[order]
hq_values = hq_values[order]


print(max(hq_radii))
hq_abs_forces = [np.linalg.norm(a) for a in hq_values[:, [3,4,5]]]
if True:
    for softening, filename, export_name in files:
        values = np.genfromtxt(filename, delimiter=',')


        # compute the radius
        radii = [np.linalg.norm(a) for a in values[:, [0,1,2]]]
        abs_forces = [np.linalg.norm(a) for a in values[:, [3,4,5]]]

        # plot the theoretical values
        plt.plot(hq_radii, hq_abs_forces, label='Hernquist', color='red', linestyle=':', zorder=1)

        plt.scatter(radii, abs_forces, s=1, label=f'softening = {softening}', zorder=0)
        plt.loglog()
        plt.xlabel('Distance to center [a.u.]')
        plt.ylabel('Absolute force value [a.u.]')
        plt.title(f'Direct Force Calculation | softening = {softening}')
        plt.legend()
        plt.savefig(export_name, dpi=1000)
        plt.close()
```
